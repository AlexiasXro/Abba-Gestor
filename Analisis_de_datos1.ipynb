## ETL Extración, Tranformación y Carga




## 1.   Extración




```python

from google.colab import drive
import pandas as pd
import chardet #libreria para los encoding

#Mostramos Google Drive para acceder al archivo
drive.mount('/content/drive', force_remount=True)

ruta_archivo = '/content/drive/MyDrive/Analisis de datos 2025/ventas.csv'

with open(ruta_archivo, 'rb') as f:
  #podemos limitar los kbites f.read(5000)
  result = chardet.detect(f.read())


try:
  df = pd.read_csv(ruta_archivo, encoding=result['encoding'], sep=';', decimal=',')

  print('Extración de datos exitosa')
except FileNotFoundError:
  print('El archivo no se encuentra en la ruta especificada')


```


## 2.   Transformación



```python
# METODO 1
# Muestra las primeras 5 filas del DataFrame
df.head(10)
```

```python
#METODO 3
#Generar estádisticas descriptivas de la columnas numericas
#count -cantidad de registros
# mean -media(valis del medio)
# std - desviacion estandar
# min - valor minimo
# 25%, 50%, 75% -cuartiles
# max
# promedio (la suma de todo los ..)
df.describe()
```

```python
# Renombrar columnas para facilitar el acceso  y evitar caracteres especiales
df.rename(columns ={
    'Tienda': 'tienda',
    'Marca' : 'marca',
    'Tipo' : 'tipo',
    'Gï¿½nero' : 'genero',
    'Talla' : 'talla',
    'Color' : 'color',
    'Categoria' : 'categoria',
    'precio de venta' : 'precio',
    'Fecha' : 'fecha',
    'Hora' : 'hora'
}, inplace=True)
```

```python


# acceder a
# .astype(str) fuerza a que sea string
df['precio'] = df['precio'].astype(str)
# remplasa la ',' por el '.'
df['precio'] = df['precio'].str.replace(',', '.',regex=True).astype(float)
#Limpieza y unificacion de los datos de la columna 'genero'
df['genero'] = df['genero'].str.replace('Masculino', 'M', regex=False).str.replace('Femenino', 'F', regex=False).str.replace('Usinex', 'U', regex=False)
print(set(df['genero']))

#Crear una columna de fecha y hora combinada y convertila al tipo datatime
df['fecha_hora'] = pd.to_datetime(df['fecha'] + ' ' + df['hora'], format='%d/%m/%Y %H:%M:%S')

# OJO(orden de ejecucion)-Eliminar las columnas originales de fecha y hora
df.drop(columns=['fecha', 'hora'], inplace=True)


print("Tranformacion de datos exitoso")
```
